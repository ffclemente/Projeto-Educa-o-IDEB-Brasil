```python
# Agora temos a base completa com dados de 2019, 2021 e 2023, contudo o interesse é a análise do IDEB entre SP e PE
# Vamos criar uma coluna que classifique as notas do IDEB entre 'Bom', 'Regular' e 'Ruim'
# Conforme o INEP, considera-se 'Bom' notas iguais ou maiores do que 6. Sendo assim, vamos classificar dessa maneira

import numpy as np

# Definindo as condições
condicoes = [
    (dados_brasil_colunas_nova['IDEB_2019'] >= 0) & (dados_brasil_colunas_nova['IDEB_2019'] <= 3.9),
    (dados_brasil_colunas_nova['IDEB_2019'] >= 4) & (dados_brasil_colunas_nova['IDEB_2019'] <= 5.9),
    (dados_brasil_colunas_nova['IDEB_2019'] >= 6) & (dados_brasil_colunas_nova['IDEB_2019'] <= 10)
]

condicoes_2 = [
    (dados_brasil_colunas_nova['IDEB_2021'] >= 0) & (dados_brasil_colunas_nova['IDEB_2021'] <= 3.9),
    (dados_brasil_colunas_nova['IDEB_2021'] >= 4) & (dados_brasil_colunas_nova['IDEB_2021'] <= 5.9),
    (dados_brasil_colunas_nova['IDEB_2021'] >= 6) & (dados_brasil_colunas_nova['IDEB_2021'] <= 10)
]

condicoes_3 = [
    (dados_brasil_colunas_nova['IDEB_2023'] >= 0) & (dados_brasil_colunas_nova['IDEB_2023'] <= 3.9),
    (dados_brasil_colunas_nova['IDEB_2023'] >= 4) & (dados_brasil_colunas_nova['IDEB_2023'] <= 5.9),
    (dados_brasil_colunas_nova['IDEB_2023'] >= 6) & (dados_brasil_colunas_nova['IDEB_2023'] <= 10)
]

# Definindo os valores para cada condição
valores = ['Ruim', 'Regular', 'Bom']

# Criando a nova coluna com base nas condições
dados_brasil_colunas_nova['IDEB_avaliacao_2019'] = np.select(condicoes, valores, default='Valor inválido')
dados_brasil_colunas_nova['IDEB_avaliacao_2021'] = np.select(condicoes, valores, default='Valor inválido')
dados_brasil_colunas_nova['IDEB_avaliacao_2023'] = np.select(condicoes, valores, default='Valor inválido')

# Exibindo as primeiras linhas para verificar
dados_brasil_colunas_nova[['IDEB_2019', 'IDEB_avaliacao_2019']].head()
dados_brasil_colunas_nova.head()
```

```python
# Agrupando os municípios conforme notas no IDEB (Bom, Regular, Ruim e Péssimo)
# Agora irei dividir a categoria 'Ruim' entre 'Ruim' e 'Péssimo', para destacar as escolas que tiveram avaliação menor do que 3.

# Converter a coluna para numérico (float)
dados_municipios_municipal['IDEB_2019'] = pd.to_numeric(dados_municipios_municipal['IDEB_2019'], errors='coerce')

# Definir as condições
condicoes = [
    (dados_municipios_municipal['IDEB_2019'].isna()),  # Condição para valores NaN
    (dados_municipios_municipal['IDEB_2019'] >= 0) & (dados_municipios_municipal['IDEB_2019'] <= 2.9),  # Péssimo
    (dados_municipios_municipal['IDEB_2019'] >= 3) & (dados_municipios_municipal['IDEB_2019'] <= 3.9),  # Ruim
    (dados_municipios_municipal['IDEB_2019'] >= 4) & (dados_municipios_municipal['IDEB_2019'] <= 5.9),  # Regular
    (dados_municipios_municipal['IDEB_2019'] >= 6) & (dados_municipios_municipal['IDEB_2019'] <= 10)   # Bom
]

# Definir resultados correspondentes às condições
resultados = ['Desconhecido', 'Péssimo', 'Ruim', 'Regular', 'Bom']

# Criar coluna 'IDEP_avaliacao_2019' com função select do pacote numpy
dados_municipios_municipal['IDEB_avaliacao_2019'] = np.select(condicoes, resultados, default='Valor Inválido')
dados_municipios_municipal.head(10)
```
